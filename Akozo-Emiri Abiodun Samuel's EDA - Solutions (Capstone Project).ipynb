```python
#6.Display the number of null values in the dataset and the total count of it
nulls_per_column = df.isnull().sum()
print("Null values per column:\n", nulls_per_column)
print("Total null values in the dataset:", nulls_per_column.sum())
```

```python
#7.Identify the columns having null values and remove them. Assume cleaned data is named as @penguins_cleaned

null_columns = df.columns[df.isnull().any()] #This finds the columns that contain any null values
penguins_cleaned = df.dropna(subset = null_columns) #This deletes the columns that contain null values.
print(penguins_cleaned)
```

```python
# 8.check the size of dataset after cleaning and compare with the size before cleaning

print (penguins_cleaned.shape) #To check the dataset size after cleaning
print (df.shape) #To confirm the dataset size before cleaning

print ("The dataset original size was", df.shape, "but reduced to", penguins_cleaned.shape, "after cleaning.")
```

```python
#9.Find out the count of male and female penguins

sex_counts = penguins_cleaned["sex"].value_counts() #Counts the values under the sex column
print(sex_counts)
```

```python
#10.Find out the count of species 

species_count = penguins_cleaned["species"].value_counts()
print(species_count)
```

```python
#11. Use a countplot graph to display the number of male and female penguins
#Give the plot a title "Count of male and female penguins"
#Give a xlabel and ylabel 

sns.countplot(data = penguins_cleaned, y = "sex", hue = "sex", palette = "Set2", legend = False)
plt.title("Count of male and female penguins")
plt.xlabel("Count")
plt.ylabel("Sex")

plt.show()
```

```python
#14. Create a histogram with the flipper_length_mm and bins = 10

sns.histplot(df["flipper_length_mm"], bins = 10, kde = False, color = "orange")
plt.title("Histogram for Flipper Length (mm)")
plt.xlabel("Flipper Length (mm)")
plt.ylabel("Frequency")
plt.show()
```

```python
#15. Create a joint point with body_length_mm, body_depth_mm for penguins_cleaned dataset

#I discovered that there are no columns that indicate body_length_mm and body_depth_mm 
#in both the original penguins dataset and the penguins_cleaned dataset,
#so I'll work with bill_length_mm and bill_depth_mm
sns.jointplot(data = penguins_cleaned, x = "bill_length_mm", y = "bill_depth_mm", kind = "scatter", color = "blue")

plt.show()
```

```python
#16. Use a pairwise plot to display all the numerical values of the dataset penguins_cleaned

numeric_columns = penguins_cleaned.select_dtypes(include = "number")
sns.pairplot(numeric_columns, corner=True)

plt.show()
```

```python
#17.Check if there exists any outliers based on the boxplot based on numerical values  of penguins_cleaned dataset

#Selecting only numeric columns from the cleaned dataset
numeric_cols = penguins_cleaned.select_dtypes(include = "number").columns


plt.figure(figsize = (14, 8))
sns.boxplot(data = penguins_cleaned[numeric_cols], orient = "h", palette = "Set3")

#Title
plt.title("Boxplot of Numerical Features in penguins_cleaned (Outlier Check)")

plt.show()
```

```python
#18. Create a violin plot to display the body_mass_g of different species based on sex

plt.figure(figsize=(10, 6))
sns.violinplot(data = penguins_cleaned, x = "species", y = "body_mass_g", hue = "sex", split = True, palette = "Set2")

#Adding title and axis labels
plt.title("Body Mass Distribution by Species and Sex")
plt.xlabel("Species")
plt.ylabel("Body Mass (g)")

plt.show()
```
